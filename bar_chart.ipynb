```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

csfont = {'fontname':'Calibri'}
```

```python
x = np.loadtxt("mips_with_comp.txt", usecols=(0))
y = np.loadtxt("mips_with_comp.txt", usecols=(1))

x1 = np.loadtxt("mips_pure.txt", usecols=(0))
y1 = np.loadtxt("mips_pure.txt", usecols=(1))

plt.figure(figsize=(10,7), frameon=False)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.bar(x1, y1, color='C2', label="All MIP papers")
plt.bar(x, y, color='C0', label="Computational MIP papers")

plt.xlabel("Publication year", fontsize=16, labelpad=11, **csfont)
plt.xticks(x, rotation=270, fontsize=12, **csfont)

plt.ylabel("Number of publications", fontsize=16, **csfont)
plt.yticks(fontsize=12, **csfont)

#plt.title("Total number of MIP publications", fontsize=20, **csfont)

plt.grid(linestyle='--', alpha=0.5)

plt.legend(fontsize=16)

plt.savefig('mips_vs_comp2.jpeg', dpi=500)
```
